__Nemenyi Test__

```python
import Orange
import matplotlib.pyplot as plt
names = ["NBE", "kNNE", "DTE", "RF", "SVME", "NNE", "NBhte", "kNNhte", "DThte", "SVMhte", "NNhte", "HTEsm", "HTEdf" ]
avranks =  [11.85, 7.10, 10.30, 6.75, 7.15, 7.10, 10.85, 7.0, 7.55, 4.85, 5.20, 3.60, 1.70]
cd = Orange.evaluation.compute_CD(avranks, n=10, alpha='0.05', test="nemenyi") #tested on 10 datasets
Orange.evaluation.graph_ranks(avranks, names, cd=cd, width=15, textspace=1.5, reverse=True)

print('CD value:', cd)
plt.show()
```
